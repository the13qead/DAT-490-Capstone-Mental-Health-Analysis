```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np




# Load the dataset
df = pd.read_csv('Mental Health dataset1.csv')

print("All Columns in the dataset (Potential Independent Variables):")
print(df.columns.tolist())

# Potential Dependent Variable for most analyses: 'Treatment'
print("\nPrimary Potential Dependent Variable: Treatment")


# Prepare data for the relationship graph (cross-tabulation of the two variables)
relationship_data = pd.crosstab(df['FamilyHistory'], df['Treatment'])

# Plotting the relationship
plt.figure(figsize=(8, 6))
# Use a stacked bar chart to compare counts across both FamilyHistory and Treatment
relationship_data.plot(kind='bar', stacked=False, ax=plt.gca())

plt.title('Relationship between Family History and Treatment')
plt.xlabel('Family History of Mental Illness')
plt.ylabel('Count of Respondents')
plt.xticks(rotation=0) # Keeps labels horizontal
plt.legend(title='Treatment Sought')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Create cross-tabulation
crosstab = pd.crosstab(df['IncreasingStress'], df['Treatment'])

# Plot stacked bar chart
crosstab.plot(kind='bar', stacked=True)
plt.xlabel('Increasing Stress Level')
plt.ylabel('Count')
plt.title('Increasing Stress vs Treatment (Stacked Bar Chart)')
plt.show()
```
